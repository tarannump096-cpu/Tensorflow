```python
import tensorflow as tf
inputs=tf.Variable([[2.3]])
weight=tf.Variable([[0.0, 0.1]])
bias=tf.Variable([[1.0]])
w=tf.matmul(inputs,weight)
res=w+bias
tf.nn.leaky_relu(res)
```
